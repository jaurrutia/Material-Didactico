\begin{eqnarray} 
\phi (\vec r) &=& k \frac{1}{||\vec r - \vec r_m||} \\
\vec E (\vec r) &=& k \frac{(\vec r - \vec r_m)}{||\vec r - \vec r_m||^3} 
\end{eqnarray}

```python
def campo(x,y,x_m=0,y_m=0):
    norma=sqrt(x**2+y**2) 
    
    E_x=x-x_m
    E_y=y-y_m
    #E_x,E_y=E_x/norma,E_y/norma#Esto hace a el campo unitario
    
    return E_x,E_y

campo(1,-1,x_m=0,y_m=0)
```

```python
x=linspace(-3,3,20)
y=linspace(-3,3,20)
X,Y=meshgrid(x,y)   #Este hace la mallita :3 (Matriz de puntos)

vec_x,vec_y=campo(X,Y,0,1)  #(0,1) es el nuevo oirgen
campo_vectorial=quiver(X,Y,vec_x,vec_y) #origen y luego direccion
```

El modelo de Lodka-Volterra consiste en un modelo log\'{i}stico de la poblaci\'{o}n de conojes y los de lobos.

\begin{eqnarray}
\dot x &=& ax+bxy\\
\dot y &=& cy+dxy
\end{eqnarray}

en donde $a>0$ y $b,c<0$. Esto ense\~{n}a c\'{o}mo cambian las poblaciones de los animales en funcio\'{o}n de los otros. Se espera que se algo medio c\'{i}clico.

```python
def Lodka_Volterra (x, t):
    a,b,c,d=0.5,-0.3,0.2,0.3
    return a*x[0]+b*x[0]*x[1],c*x[0]+d*x[0]*x[1]
```

```python
lobos=linspace(0,5,20)
conejos=linspace(0,2.5,20)
C,L=meshgrid(conejos,lobos)
C_dot,L_dot=Lodka_Volterra([C,L],0)
campo_vectorial= quiver(C,L,C_dot,L_dot)
```
