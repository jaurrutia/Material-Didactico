```python
def borrachito(n):
    pos=[] 
    paso=0
    for i in range(n):
        volado=rand()
        if volado>0.5:
            paso=paso+1
            pos.append(paso)
        else:
            paso=paso-1
            pos.append(paso)
    return pos
```

```python
pinches_vagos=[]
for i in range(10):
    pinches_vagos.append(borrachito(10000))
for i in range(10):
    plot(pinches_vagos[i])
```
