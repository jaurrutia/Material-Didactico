```python
from numpy import *
from math import *
from scipy import *
from scipy.integrate import *
from pylab import *
%matplotlib inline
```

\begin{equation} V(x,y)=\sin(x)\cos(y) \end{equation}

```python
def potencial(x,y):
    return sin(x)*cos(y)
```

```python
x=linspace(-2*pi,2*pi,200)
y=x
X,Y=meshgrid(x,y)

z=potencial(X,Y)
contour(X,Y,z)  #Traza equipotenciales
```

```python
contourf(X,Y,z)  #Traza equipotenciales con cdigos de color
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig=figure()
a=fig.add_subplot(1,1,1,projection='3d')
figurita = a.plot_surface(X,Y,z,cmap=cm.coolwarm)
```

```python
from numpy import *
from math import *
from scipy import *
from scipy.integrate import *
from pylab import *
from mpl_toolkits.mplot3d.axes3d import Axes3D
%matplotlib inline

def potencial(x,y):
    return sin(x)*cos(y)

x=linspace(-2*pi,2*pi,200)
y=x
X,Y=meshgrid(x,y)
z=potencial(X,Y)

fig=figure()
a=fig.add_subplot(1,1,1,projection='3d')
figurita = a.plot_surface(X,Y,z,cmap=cm.coolwarm)
```
