$$
y'' + \omega_0^2\sin y =g(t)\\
y' = u\\
u'= -\omega_0^2 \sin y + g(t)
$$

```python
from pylab import *
%run "C:\Users\Jonathan\Documents\GitHub\integradores\int_rk_jac.ipynb"   #Importo las funciones que use aqu√≠
%matplotlib inline
```

```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])])
```

```python
t,sol = rk4(osc,[1,0.],[0,10],.1)
zoom = 2
xlim([-zoom,zoom]),ylim([-zoom,zoom])  #El equiva√±ente a Range y Domain en Mathematica
axes().set_aspect('equal'),grid(True)   #La maquilla
plot(sol[:,0],sol[:,1])    #Me da es espacio fase
show()
```

Ahora vamos a modifir la soluc√≥n para que la g(t) no sea sero (como en el caso pasado)
$$g(t) = 0.1 \sin t$$

```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])+0.1*sin(t)])
```

```python
for i in range(3):   
    y0, u0 = pi*(2*rand()-1), 0   #Genera las cosasas aleatorias  rand()-> N√∫mero aleatorio entre 0 y 1
    t,sol = rk4(osc,[y0,u0],[0,500])
    plot(sol[:,0],sol[:,1])    #Me da es espacio fase en cada ciclo
    
zoom = pi
xlim([-zoom,zoom]),ylim([-zoom,zoom])  #El equiva√±ente a Range y Domain en Mathematica
axes().set_aspect('equal'),grid(True)   #La maquilla
show()
```
