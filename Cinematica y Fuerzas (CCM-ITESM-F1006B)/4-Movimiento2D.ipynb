## Cantidades dinámicas

Para describir el movimiento de una partícula, se emplea el vector posición $\vec{r}$, el cuál da una expresión para conocer en qué punto del espacio tridimencional se encuentra la partícula como función del tiempo $t$.

Es posible calcular los cambios de la posición al pasar el tiempo *calculando la derivada respecto al tiempo* de esta canidad, lo que determina la antidad física llamada *velocidad instantánea*. Al volver a derivar, se obtiene la relación con la *aceleración*.

En coordenadas cartesianas, la derivación respecto a tiempo se puede realizar componente a componente, es decir, se cumplen las siguientes expresiones:

   $$
   \vec{r} =  (x (t),\; y(t),\; z(t))  \qquad\Longrightarrow  \qquad\vec{v} =  \dot{\vec{r}} =  (\dot{x} (t),\; \dot{y}(t),\; \dot{z}(t))  
   \qquad\Longrightarrow   \qquad\vec{a} =  \ddot{\vec{r}} =  (\ddot{x} (t),\; \ddot{y}(t),\; \ddot{z}(t))  
   $$


### Combinando dos situaciones conocidas

Dado que en coordenadas cartesianas es posible realizar las derivaciones respecto al tiempo componente a componente, consideramos anteriormente el caso unidimensional para analizar los casos donde:

 - La velocidad es constante: **Movimiento rectilíneo uniforme (MRU)**
 - La aceleración es constante: **Movimiento uniformemente acelerado (MUA)**

Dado que en coordenadas cartesianas cada componente se comporta de forma independiente de las demás, basta con identificar el tipo de movimiento en cada coordenada y recordar que ahora estamos **parametrizando la trayectoria** como función de un parámetro temporal $t$.

## Tiro parabólico

En este caso bidimesional consideraremos que no la única fuerza que actúa sobre una partícula es la de la gravedad, que siempre actúa en dirección al centro de la Tierra. Como consecuencia de esto, en la componente vertical se tiene un MUA con la aceleración $a_y = -g$, mientas que en la componente horizontal un MRU

### Expresiones notables

En notación vectorial el tiro parabólico se escribe como

$$
\vec{r}(t) = \vec{v}_0 (t - t_0) + \frac12 \vec{a}(t-t_0)^2,
$$
donde para la componente horizontal se cumple que
$$
x(t) = x_0 + v_{0,x} (t -t_0), \qquad\text{con}\qquad v_x(t) = v_{0,x} = ||\vec{v}_0|| \cos\theta_0.
$$
con $\theta_0$ el ángulo de la velocidad al tiempo $t_0$, y donde $a_x = 0$. Por otro lado, para la componetne vertical se tiene que
$$
y(t) = y_0 + v_{0,y} (t -t_0) - \frac12 g (t - t_0)^2, \qquad\text{con}\qquad v_y(t) = v_{0,y}- g (t-t_0),\qquad\text{y}\qquad v_{0,y} = ||\vec{v}_0|| \sin\theta_0.
$$

Cabe mencionar que con las expresiones anteriores se observa que la dirección y la magnitud del vector velocidad cambia en el tiempo. Sin emabrgo, aún es posible emplear las expresiones para la norma y dirección del vector (con el ángulo medido respecto al eje horizontal), sólo hay que recordad que todos los parámetros dependen del tiempo:

$$
||\vec{v}(t)||^2 = v_x^2(t) + v_y^2(t), \qquad\qquad \tan[\theta(t)] = \frac{v_y(t)}{v_x(t)}.
$$
