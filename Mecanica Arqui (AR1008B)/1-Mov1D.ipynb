### Simplificando en una sola dimensión

En esta sesión estudiaremos el caso de un objeto que se mueve en **linea recta**. Abordaremos los casos más sencillos primero:

 - La velocidad es constante: **Movimiento rectilíneo uniforme (MRU)**
 - La aceleración es constante: **Movimiento uniformemente acelerado (MUA)**

Finalmente, aprenderemos a **leer gráfivas de posición vs. tiempo**. Con estas semremos capaces de entender cualquier movimiento unidimensional.


## Movimiento Rectilíneo Uniforme (MRU)

En el caso unidimensional podemos considerar a la coordenada $x(t)$ como la que describe a una partícula que puede moverse a lo largo de una línea recta. 

Si consideramos el caso en el que la **velocidad es constante**, entonces ésta es igual a la velocidad promedio, por lo que se puede despejar la posición de la siguiente manera:

$$
v = \frac{\Delta x(t)}{\Delta t} = \frac{x(t)-x_0}{t-t_0} \qquad\Longrightarrow\qquad x(t) = x_0 + v (t - t_0)
$$

Lo que corresponde a la ecuaciónd e una recta con oordenda al origen $x_0$, pendiente $v$ y desplazada $t_0$ unidades a la derecha.

```
<<"MaTeX`"
texStyle = { FontFamily -> "Latin Modern Roman", FontSize -> 12,Black};
TeXiT = Style[#,texStyle]&;

framext = (MaTeX[#,FontSize->13]&/@{"t\\text{ [s]}","x(t)\\text{ [m]}"});
framevt = (MaTeX[#,FontSize->13]&/@{"t\\text{ [s]}","v(t)\\text{ [m]}"});
frameat = (MaTeX[#,FontSize->13]&/@{"t\\text{ [s]}","a(t)\\text{ [m]}"});

x[t_]:= x0 + v * (t - t0)
time = Range[0,10,.1];
```

## Movimiento Uniformemente Acelerado (MUA)

De nuevo considerando el caso unidimensional, podemos estudiar la dinámica del sistema en un caso sencillo.

Si consideramos el caso en el que la **aceleración es constante**, entonces ésta es igual a la aceleración promedio, por lo que se puede despejar la velocidad* de la siguiente manera:

$$
a = \frac{\Delta v(t)}{\Delta t} = \frac{v(t)-v_0}{t-t_0} \qquad\Longrightarrow\qquad v(t) = v_0 + a (t - t_0),
$$

lo que corresponde a la ecuación de una recta con ordenda al origen $v_0$, pendiente $a$ y desplazada $t_0$ unidades a la derecha. En el caso de la posición tenemos

$$
x(t) = x_0 +  v_0 (t-t_0) + \frac{1}{2} a (t - t_0)^2,
$$

es decir, que el movimiento en este caso está descrito por una parábola.

## Ejemplos:

 - Se lanza una pelota hacia arriba con masa $m$ a una velocidad de $5$ m/s. ¿Cuánto tiempo tarda la pelota al suelo en llegar a la altura de la que fue lanzada?
 - ¿Cuál es la altura máxima que alcanza la pelota?
 - Si una segunda pelota se lanza sobre la misma linea, 0.5 segundos después de la primera con una velocidad de $4$ m/s ¿Encuanto tiempo y a qué altura se encuentran las pelota?

```
pos[t_]:= x0 + v0 * (t-t0) + .5 * a * (t - t0)^2
img = Import["https://upload.wikimedia.org/wikipedia/commons/2/25/Free-fall.gif"];
time = Range[0,Length[img]-1]/Length[img];

position = Table[{t,(x0 = 0; v0 = 0; t0 = 0;
                        a = -9.81;pos[t]) },{t,time}];
exp = Table[Row[{Show[img[[i]],ImageSize->150],
        ListPlot[position[[;;i]],
            PlotTheme -> "Scientific", FrameLabel -> framext, Joined->True,Mesh->All,
            FrameStyle -> Black,AspectRatio -> 1,PlotRange->{{0,1},{-5.25,1}},
            Frame -> True, ImageSize -> 350, 
            BaseStyle->texStyle,
            PlotLabel->TeXiT["Pelota que se deja caer"]]
        }, BaselinePosition  -> Top],
    {i,Length[time]}];


Export[FileNameJoin[{"img","FreeFall.gif"}],exp,"DisplayDurations" -> .075(*seconds*)]
```

```
pos[t_]:= x0 + v0 * (t-t0) + .5 * a * (t - t0)^2

Plot[(
    x0 = 0; v0 = 5; t0 = 0;
    a = -9.81;
    pos[t]),
        {t,0,1.025},  PlotTheme -> "Scientific", FrameLabel -> framext, 
        FrameStyle -> Black,
        Frame -> True, ImageSize -> 500, 
        BaseStyle->texStyle,
        PlotLabel->TeXiT["Pelota lanzada hacia arriba"]]
```

```
Plot[{
    (x0 = 0; v0 = 5; t0 = 0;
    a = -9.81;
    pos[t]),
    
    (x0 = 0; v0 = 4; t0 = .5;
    a = -9.81;
    pos[t])
    },
        {t,0,1.025},  PlotTheme -> "Scientific", FrameLabel -> framext, 
        FrameStyle -> Black,PlotRange->{0,1.2},
        Frame -> True, ImageSize -> 500, 
        BaseStyle->texStyle,
        PlotLabel->TeXiT["Dos pelotas encontrándose"]]
```
